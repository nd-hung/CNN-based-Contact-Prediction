```python
# model evaluation on Psicov150 test set
import os
import subprocess
from datetime import datetime

# path to ground truth
pdb_dir = './data/DeepCov/test/psicov150/pdb'

# path to predicted results
#pred_dir = './data/DeepCov/test/psicov150/rr'
pred_dir = './data/DeepCov/test/psicov150/rr'

# test target list
target_file = './data/DeepCov/test/psicov150/target.lst.sorted'


# get total number of protein sequences
num_prots = len(open(target_file, "r").readlines())
count = 0   # for progress display

date_time = datetime.now().strftime("%Y%h%d_%H%M")
result_file = "result/deepcon_" + date_time + ".csv"

if os.path.isfile(result_file):
   !rm result_file
else:
    with open(target_file, 'r') as f:
        for line in f:
            target = line.rstrip()
            ss = os.path.join(pdb_dir, target + '.pdb')
            ts = os.path.join(pred_dir, target + '.rr')
            if os.path.isfile(ss) and os.path.isfile(ts):
                p = subprocess.Popen(["perl", "./coneva-lite3.pl", 
                                      "-pdb", ss, 
                                      "-rr", ts,
                                      "-o", result_file])
            else: 
                print('File(s) not found.')
                pass
            count += 1
            print('{}/{} - {}'.format(count, num_prots, target))

print('Evaluation done.')
```
