```python
# Extract alignment to .c21 covariance matrix using cov21stats
# Pls run setup.sh script first to build the cov21stats excecutable file (stored in ./bin by default)
import os

# set path to aln directory and feature directory

# paths for extracting training data
train_aln_dir = './data/train/aln/'
train_feat_dir = './data/train/21c/'
train_target_file = '.data/train/train_ecod.lst'

# paths for extracting test data
test_aln_dir = './data/test/psicov150/aln/'
test_feat_dir = './data/test/psicov150/21c/'
test_target_file = './data/test/psicov150/target.lst.sorted'

# feature extraction for training data
# get total number of protein sequences
num_prots = len(open(train_target_file, "r").readlines())
count = 0 # for progress report

with open(train_target_file, 'r') as f:
    for line in f:
        target = line.rstrip()
        ss = train_aln_dir + target + '.aln'
        ts = train_feat_dir + target + '.21c'
        if os.path.isfile(ss):
            if not os.path.isfile(ts):
                cmd = 'bin/cov21stats ' + ss + ' ' + ts
                os.system(cmd)
            else: # feature file exists
                # print('Feature file {} existed. Skip.'.format(ts) )
                continue
        else: # aln file does not exist
            # print('Source file {} does not exist. Skip.'.format(ss) )
            continue
        count += 1
        print('{}/{} extracted.'.format(count, num_prots))

# feature extraction for test data
# get total number of protein sequences
num_prots = len(open(test_target_file, "r").readlines())
count = 0 # for progress report

with open(test_target_file, 'r') as f:
    for line in f:
        target = line.rstrip()
        ss = test_aln_dir + target + '.aln'
        ts = test_feat_dir + target + '.21c'
        if os.path.isfile(ss):
            if not os.path.isfile(ts):
                cmd = 'bin/cov21stats ' + ss + ' ' + ts
                os.system(cmd)
            else: # feature file exists
                # print('Feature file {} existed. Skip.'.format(ts) )
                continue
        else: # aln file does not exist
            # print('Source file {} does not exist. Skip.'.format(ss) )
            continue
        count += 1
        print('{}/{} extracted.'.format(count, num_prots))
```
